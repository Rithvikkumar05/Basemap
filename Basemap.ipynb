```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
fig = plt.figure(figsize=(8,8))
m=Basemap()
m.drawcoastlines()
```

```python
m= Basemap(projection = 'ortho',lat_0=0,lon_0=0)
m.drawmapboundary(fill_color ='aqua')
m.fillcontinents(color = 'coral',lake_color='aqua')
m.drawcoastlines()
plt.show()
```

```python
fig=plt.figure(figsize=(8,8))
m = Basemap(projection = 'ortho',lat_0=50,lon_0=0)
m.drawmapboundary(fill_color ='aqua')
m.fillcontinents(color = 'coral',lake_color='aqua')
m.drawcoastlines()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8),edgecolor = 'w')
m= Basemap(projection = 'cyl',
            llcrnrlat=-40, urcrnrlat = 40,
            llcrnrlon=-60, urcrnrlon= 60)
m.drawcoastlines()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8),edgecolor = 'w')
m= Basemap(projection = 'cyl',
            llcrnrlat=-40, urcrnrlat = 40,
            llcrnrlon=-60, urcrnrlon= 60)
m.shadedrelief(scale = 0.99)
```

```python
fig =plt.figure(figsize=(8,8),edgecolor = 'w')
m = Basemap(projection = 'lcc',height = 8E6,
            width =8E6,lat_0 = 20.5937,lon_0 = 78.9629)
m.etopo(scale = 0.8,alpha = 0.5)

x,y = m(77.0797,10.99979)
plt.plot(x,y,'ok',markersize = 5)
plt.text(x,y, 'hello',fontsize = 12);
```
